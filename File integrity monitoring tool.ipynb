```python

import os
import hashlib
import tkinter as tk
from tkinter import filedialog, messagebox

def generate_sha256(file_path):
    """
    Generates SHA-256 hash of a given file.
    """
    sha256 = hashlib.sha256()

    try:
        with open(file_path, "rb") as file:
            while True:
                chunk = file.read(4096)
                if not chunk:
                    break
                sha256.update(chunk)
        return sha256.hexdigest()

    except Exception as e:
        messagebox.showerror("Error", f"Unable to read file.\n{e}")
        return None

class FileIntegrityApp:

    def __init__(self, root):
        self.root = root
        self.root.title("File Integrity Monitoring Tool")
        self.root.geometry("650x420")
        self.root.configure(bg="#1e1e2f")
        self.root.resizable(False, False)

        self.selected_file = ""
        self.baseline_hash = ""

        self.create_widgets()

    def create_widgets(self):

        title = tk.Label(
            self.root,
            text="File Integrity Monitoring Tool",
            font=("Arial", 18, "bold"),
            bg="#1e1e2f",
            fg="white"
        )
        title.pack(pady=15)

        browse_btn = tk.Button(
            self.root,
            text="Select File",
            command=self.select_file,
            width=20,
            bg="#4CAF50",
            fg="white",
            font=("Arial", 11, "bold")
        )
        browse_btn.pack(pady=5)

        self.file_label = tk.Label(
            self.root,
            text="No file selected",
            wraplength=600,
            bg="#1e1e2f",
            fg="#cccccc"
        )
        self.file_label.pack(pady=5)

        hash_title = tk.Label(
            self.root,
            text="Baseline SHA-256 Hash:",
            font=("Arial", 11, "bold"),
            bg="#1e1e2f",
            fg="white"
        )
        hash_title.pack(pady=5)

        self.hash_display = tk.Label(
            self.root,
            text="-----",
            wraplength=600,
            bg="#1e1e2f",
            fg="#00ffcc"
        )
        self.hash_display.pack(pady=5)

        check_btn = tk.Button(
            self.root,
            text="🔍 Verify Integrity",
            command=self.verify_file,
            width=20,
            bg="#2196F3",
            fg="white",
            font=("Arial", 11, "bold")
        )
        check_btn.pack(pady=15)

        reset_btn = tk.Button(
            self.root,
            text="Reset",
            command=self.reset_tool,
            width=20,
            bg="#f44336",
            fg="white"
        )
        reset_btn.pack(pady=5)

        self.status_label = tk.Label(
            self.root,
            text="Status: Waiting for file selection",
            font=("Arial", 12),
            bg="#1e1e2f",
            fg="yellow"
        )
        self.status_label.pack(pady=15)

        footer = tk.Label(
            self.root,
            text="Developed by Amara Afzal | BS Cyber Security",
            font=("Arial", 9),
            bg="#1e1e2f",
            fg="#888888"
        )
        footer.pack(side="bottom", pady=10)

    def select_file(self):
        self.selected_file = filedialog.askopenfilename()

        if self.selected_file:
            self.file_label.config(text=self.selected_file)

            self.baseline_hash = generate_sha256(self.selected_file)

            if self.baseline_hash:
                self.hash_display.config(text=self.baseline_hash)
                self.status_label.config(
                    text="Baseline hash generated successfully",
                    fg="lightblue"
                )

    def verify_file(self):
        if not self.selected_file:
            messagebox.showerror("Error", "Please select a file first!")
            return

        current_hash = generate_sha256(self.selected_file)

        if current_hash == self.baseline_hash:
            self.status_label.config(
                text=" File is SAFE (No changes detected)",
                fg="lightgreen"
            )
        else:
            self.status_label.config(
                text="WARNING: File has been modified!",
                fg="red"
            )

    def reset_tool(self):
        self.selected_file = ""
        self.baseline_hash = ""
        self.file_label.config(text="No file selected")
        self.hash_display.config(text="-----")
        self.status_label.config(
            text="Status: Waiting for file selection",
            fg="yellow"
        )

if __name__ == "__main__":
    root = tk.Tk()
    app = FileIntegrityApp(root)
    root.mainloop()
```
